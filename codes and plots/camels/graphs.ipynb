```python
import h5py
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np

with h5py.File('/home/jovyan/Data/FOF_Subfind/IllustrisTNG/LH/LH_0/fof_subhalo_tab_024.hdf5', 'r') as hf:
    print (hf['Subhalo'].keys())
    mass = hf['Subhalo/SubhaloMass'][:]
    vmax = hf['Subhalo/SubhaloVmax'][:]
    wind = hf['Subhalo/SubhaloWindMass'][:]
    spin = hf['Subhalo/SubhaloSpin'][:]
```

```python
mass
```

```python
MASS = np.log10(mass*(10**10)) #10^10 solar mass,
```

```python
MASS
```

```python
vmax
```

```python
plt.scatter(MASS, vmax, c = vmax)
plt.colorbar()
plt.xlabel('Mass')
plt.ylabel('Vmax')
plt.show()
```

```python
plt.scatter(MASS, WIND, c= WIND)
plt.colorbar()
plt.xlabel('Mass')
plt.ylabel('wind')
plt.show()
```
